# Basic analysis

```python
# Show me transactions over £1m each?

data[(data['Amount'] > 1000000)].sort_values(by='Amount', ascending=False)
```

```python
# Show me all expenses associated with COVID-19

data[data['Expense Area'].str.contains("COVID")].sort_values(by="Amount", ascending=False)
```

```python
# How much did the Cabinet Office pay Accenture in May 2020?

data[data['Supplier'].str.contains("ACC") | data['Supplier'].str.contains("Acc")]
```

```python
# BONUS!

# There's a more eligant way to filter on values, using 'loc'

data.loc[data.Supplier.str.contains("Acc") | data.Supplier.str.contains("ACC")]
```

# Let's tidy the data up!

```python
# That's all well and good, but the data is kind of messy...

data.loc[data.Supplier.str.contains("Manning Gott") | data.Supplier.str.contains("MANNING GOTT")].tail(2)
```

```python
# Notice the typo? (Hopefully the Cabinet Office did)
```

```python
# Lets fix that, and the others... including: Amazon, Accenture, Capita, Interserve, and PA Consulting
```

```python
data.replace("MANNING GOTTLIEB OMD","Manning Gottlieb OMD Limited",inplace=True)
data.replace("Manning Gottlieb OMD","Manning Gottlieb OMD Limited",inplace=True)
data.replace("Manning Gottileb OMD Ltd","Manning Gottlieb OMD Limited",inplace=True)
data.replace()

data.replace("PA CONSULTING SERVICES LTD", "PA Consulting Limited", inplace=True)
data.replace("PA CONSULTING GROUP","PA Consulting Limited", inplace=True)
data.replace("AMAZON WEB SERVICES, INC.","Amazon Web Services EMEA SARL",inplace=True)
data.replace("AMAZON WEB SERVICES EMEA SARL","Amazon Web Services EMEA SARL",inplace=True)
data.replace("AWS SERVICES INC","Amazon Web Services EMEA SARL",inplace=True)
data.replace("Accenture","Accenture Limited",inplace=True)
data.replace("ACCENTURE UK LTD","Accenture Limited",inplace=True)
data.replace("CAPITA EMPLOYEE BENEFITS LTD","Capita Limited",inplace=True)
data.replace("CAPITA BUSINESS SERVICES LTD","Capita Limited",inplace=True)
data.replace("CAPITA BUSINESS SERVICES LTD TA KNOWLEDGEPOOL","Capita Limited",inplace=True)
data.replace("INTERSERVE FACILITIES MANAGEMENT LTD","Interserve Limited",inplace=True)
data.replace("Interserve (Facilities Management) Ltd","Interserve Limited",inplace=True)
```

```python
# Let's check the changes to the typo..

data[data.Supplier.str.contains("Manning Gott")].tail(2)
```

```python
# Using this newly cleansed data, show me all transactions to Amazon over £100,000

data.loc[(data['Supplier'] == "Amazon Web Services EMEA SARL") & (data['Amount'] > 100000)]
```

```python
# Let's group the data by amounts and sort them from highest to lowest
data = data.groupby(["Supplier"]).agg({"Amount": "sum"}).reset_index()
data = data.sort_values(by="Amount", ascending=False)
```

```python
data.head()
```

```python
# How many projects did the top-10 suppliers deliver to the Cabinet Office in May 2020?
data_head = data.head(10)
vals = pd.DataFrame([])

supplier_list = data_head.Supplier.tolist()
supplier_list.sort()

print("Format: supplier (number of projects)\n")
          
for i in supplier_list:     

    print(i,"(", data_og.Supplier.str.count(i).sum(),")")
```

# Tables are good, but visualisations are better!

```python
# set up the chart presentation
plt.figure(figsize=(50,25))
plt.xticks(rotation = 90, fontsize=30)
plt.yticks(fontsize=30)
plt.title("Cabinet Office Top-10 Expenses (May 2020)", fontsize=50, fontweight='bold')

# create the Seaborn chart
sns.barplot(data_head['Supplier'], data_head['Amount'], data=data_head, palette="Blues_d")
plt.ylabel("Amount (£)", fontsize=45)
plt.xlabel("Supplier (Top-10)", fontsize=45)

plt.savefig("bar.png",dpi=100, bbox_inches="tight")
```

```python
# Has any one supplier charged the Cabinet Office the most in May 2020?

top50 = data.head(50)
plt.figure(figsize=(20,5))
plt.xticks(rotation = 90, fontsize=12)
plt.annotate("Manning Gottlieb OMD Limited", (6800000,0.0012), fontsize=11.5)
plt.annotate("Cabinet Office", (4400000,0.0005), fontsize=11.5)
sns.stripplot(x=top50["Amount"], size=6, jitter=0.15, linewidth=0.6, edgecolor='black', color="yellow")
plt.ylabel("Supplier (Top-50)", fontsize=15)
plt.xlabel("Amount (£)", fontsize=15)
sns.set(style="whitegrid")


# Lets save the plot to a file
plt.savefig("scatter.png",dpi=300, bbox_inches="tight")
```
